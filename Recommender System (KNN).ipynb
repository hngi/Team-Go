```python
#At the end of this section to recommend posts to users we will use the 'all_user_post_ratings.csv'
#To recommend users we will use final_user_ratings.csv
```

```python
#Let's generate 2d matrices for both
post_recommend_matrix = all_user_ratings.pivot(index='post_id',columns='user_id',values='rating').fillna(value=0)
users_recommend_matrix = user_ratings.pivot(index='rated_user', columns='user_id_x', values='rating').fillna(value=0)

```

```python
from scipy.sparse import csr_matrix

mat_post_recommend = csr_matrix(post_recommend_matrix.values)
mat_users_recommend = csr_matrix(users_recommend_matrix.values)
```

```python
from sklearn.neighbors import NearestNeighbors

post_model = NearestNeighbors(metric='cosine',algorithm='brute')
users_model = NearestNeighbors(metric='cosine',algorithm='brute')
```

```python
post_model.fit(mat_post_recommend)
```

```python
query_index = np.random.choice(post_recommend_matrix.shape[0])
distances, indices = post_model.kneighbors(post_recommend_matrix.iloc[query_index,:].values.reshape(1,-1), n_neighbors = 6)

for i in range(0,len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}\n'.format(post_recommend_matrix.index[query_index]))
    else:
        print('{0}: {1} with a distance of {2}:'.format(i, all_user_post_ratings['title'][[indices.flatten()[i]]], distances.flatten()[i]))
        
    
```

```python
users_model.fit(mat_users_recommend)
```

```python
query_index_1 = np.random.choice(post_recommend_matrix.shape[0])
distances, indices = users_model.kneighbors(users_recommend_matrix.iloc[query_index,:].values.reshape(1,-1), n_neighbors = 6)

for i in range(0,len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}\n'.format(users_recommend_matrix.index[query_index_1]))
    else:
        print('{0}: {1} with a distance of {2}:'.format(i, user_ratings['name'][[indices.flatten()[i]]], distances.flatten()[i]))
        
    
```
