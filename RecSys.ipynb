```
# Imports
import pandas as pd

import numpy as np
```

```
# Load Lucid data
raw_user = 'https://raw.githubusercontent.com/bobby360/Team-Go/master/users.csv'
users = pd.read_csv(raw_user)
users = users.sort_values('user_id')
users.head(10)
```

```
# Load Lucid data
raw_following = 'https://raw.githubusercontent.com/bobby360/Team-Go/master/following.csv'
following = pd.read_csv(raw_following)
following = following.sort_values('follower_id')
following.head(10)
```

```
# Load Lucid data
raw_posts = 'https://raw.githubusercontent.com/bobby360/Team-Go/master/posts.csv'
posts = pd.read_csv(raw_posts)
posts = posts.sort_values('post_id')
posts.head(10)
```

```
# Load Lucid data
raw_notifications = 'https://raw.githubusercontent.com/bobby360/Team-Go/master/notifications.csv'
notifications = pd.read_csv(raw_notifications)
notifications = notifications.sort_values('sender_id')
notifications.head(10)
```

Defining function for returning recommendations on who to follow

```
indices = pd.Series(df1.index, index=df1['follower_id'])


def get_rec(follower_id):
  idx = indices[follower_id]
  friend_rec = list(enumerate(idx))
  friend_rec = sorted(friend_rec)
  #friend_rec = friend_rec[0:15]
  rec_indices = [i[0] for i in friend_rec]
  return df1[' name'].iloc[rec_indices].drop_duplicates()
# idx = indices.[follower_id]
# idx_check = 
# movie_indices = [i[0] for i in sim_scores]

# friend_rec = [i[0] for i in follower_id]
```
